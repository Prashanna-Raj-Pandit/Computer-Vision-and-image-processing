```python
import cv2
cap=cv2.VideoCapture(0)

line_start = (100, 100)
line_end = (500, 500)

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

writer=cv2.VideoWriter('capturedvideo.mp4',cv2.VideoWriter_fourcc(*'DIVX'),20,(width,height)) # 20 if FPS
while True:
    
    ret,frame=cap.read()  # tuple unpacking  
    # operations (drawing)
    writer.write(frame)
    cv2.line(frame, line_start, line_end, (0, 0, 200), 1)
   # gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1)& 0xFF==ord('q'): #ord=ordinal
        break
cap.release()
writer.release()
cv2.destroyAllWindows()
```

```python
import cv2
import time
cap=cv2.VideoCapture('capturedvideo.mp4')

if cap.isOpened()==False:
    print('ERROR FILE NOT FOUND')

while cap.isOpened():
    
    ret,frame=cap.read()
    
    if ret==True:
        # write 20 FPS
        time.sleep(1/20)
        cv2.imshow('recordong',frame)# recording is the title of window where each frame is displayed
        
        if cv2.waitKey(1)& 0xFF==ord('q'):
            break
            
            
    else:
        break
        
        
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap=cv2.VideoCapture(0)

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#top left corner
x=width//2   # // allows ro return an integer
y=height//2

# width and height of rectangle

w=width//4
h=height//4

# bottom right corner x+w, y+h

while True:
    ret,frame=cap.read()
    cv2.rectangle(frame,(x,y),(x+w,y+h),color=(0,0,255),thickness=4)
    
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()

```

```python
import cv2


# callback function
def draw_rectangle(event,x,y,flags,param):
    global pt1,pt2,topLeft_clicked,bottomRight_clicked
    
    if event==cv2.EVENT_LBUTTONDOWN:
        
        #Resrt the rectangle(it checks if the rect there)
        if topLeft_clicked==True and bottomRight_clicked==True:
            pt1=(0,0)
            pt2=(0,0)
            topLeft_clicked=False
            bottomRight_clicked=False
        if topLeft_clicked==False:
            pt1=(x,y)              
            topLeft_clicked=True
            
        elif bottomRight_clicked==False:
            pt2=(x,y)
            bottomRight_clicked=True

## global variables
pt1=(0,0)
pt2=(0,0)
topLeft_clicked=False
bottomRight_clicked=False
#connect to callback

cap=cv2.VideoCapture(0)
cv2.namedWindow('test')
cv2.setMouseCallback('test',draw_rectangle)

while True: 
    ret,frame=cap.read()
    
    #drawing on the frame based on global variables
    if topLeft_clicked:
        cv2.circle(frame,center=pt1,radius=5,color=(0,0,255),thickness=-1)
        
    if topLeft_clicked and bottomRight_clicked:
        cv2.rectangle(frame,pt1,pt2,(0,0,255),3)
    
    cv2.imshow('test',frame)
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()

```
